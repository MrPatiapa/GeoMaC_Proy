## Modelo Conceptual

Los campos geotérmicos son sistemas complejos e inherentemente dinámicos que evolucionan sobre periodos de tiempo corto. Por un lado, atraviesan cambios físicos como consecuencia de operaciones de explotación (a lo largo de años), y por otro, existen cambios que sufre el sistema geotérmico previo a su explotación como resultado de su evolución geológica (miles-millones de años). Por lo que entender y determinar el tiempo en que ocurrieron estos cambios geológicos permite la reconstrucción de la historia térmica del sistema geotérmico. Lo cual puede emplearse para determinar el potencial y tipo de un “play geotérmico “. 

Los modelos numéricos prueban que el enfriamiento de una fuente magmática debida solo a la transferencia de calor conductiva es un proceso que puede llevar millones de años. Sin embargo, la circulación hidrotermal puede generar suficiente flujo de calor convectivo para reducir drásticamente el tiempo de enfriamiento. La permeabilidad de la roca es el factor clave para determinar el mecanismo dominante de transferencia de calor en un sistema geotérmico. La permeabilidad de las rocas superpuestas sobre una fuente de calor magmático puede cambiar a través del tiempo de manera similar al régimen de flujo de calor. Cambios en la permeabilidad en sistemas geotérmicos obedecen primordialmente a interacciones roca-fluido.

Acoluco es un area prospectiva geotérmica en el estado de Puebla, México, en el cual, a la fecha, no ha progresado hacia una explotación comercial. El sistema geotérmico está alojado por un complejo de caldera volcánica que se extiende sobre la frontera de Puebla e Hidalgo, en la parte este del Cinturón Trans- Volcánico Mexicano. Este complejo es de ~18km en diámetro y contiene hasta ~900[m] de rocas volcánicas calcarinas del Plioceno al Pleistoceno, las cuales descansan sobre un basamento sedimentario del Mesozoico. Las rocas del basamento incluyen, de la base a la cima: (a) Areniscas Jurásicas, (b) Calcarenitas Cretácicas y calizas de arrecife. Asimismo, se encuentra una intrusión granítica del Cretácico tardío que meta morfea la sección sedimentaria bajo el complejo de la caldera

Los datos obtenidos son comparados con datos experimentales provistos por las inclusiones de fluidos y los registros de pozo térmicos. 

## Modelo Matemático

Para evaluar la evolución de la actividad geotérmica en Acoculco se calculo una distribución espacial 1D (vertical) de distribución de temperaturas. Desde la superficie a los 1800[m] de profundidad, lo que corresponde al intervalo de profundidad para el cual existen datos experimentales a partir del pozo exploratorio EAC-1. Se emplearon valores de T$_H$ de inclusiones de fluidos para graficar el perfil inicial (tiempo = 0), y valores de registro de temperatura para representar la geoterma actual. 

Ambas curvas son extrapoladas por el modelo hasta los 4000 [m] de profundidad, la que es la profundidad sugerida para la fuente de calor magmático. Esta extrapolación permite la integración de todas las condiciones de frontera bajo la asunsión principal de una transferencia de calor puramente conductiva, por lo cual se emplea la ecuación de balance:

$\frac{\partial T}{\partial t} = D_{th} \frac{\partial T^2}{\partial z^2}$

Donde:

$D_{th} = \frac{k}{\rho c}$

   ***Parámetros físicos***

$c_p$ : Capacidad calorifica especifica [J/KgK]

$\rho$ : Densidad [Kg/m$^3$]

$\kappa$ : Conductividad térmica [W/mK]

***Variables Independientes***

z : Profundidad [m]

t : Tiempo [s]

***Variables Dependientes***

T : Temperatura [K] 

$D_{th}$ : Difusividad térmica [m$^2$/s]


Con condiciones de forntera tipo *Dirichlet*: donde la variable dependiente ***T*** esta dada en las fronteras, siendo estas la superficie (0[m]) y la profundidad a la que se encuentra la fuente de calor magmatica (4000[m])

*T(z = 0) = T$_A$* = 15 [°C]

*T(z = 4000) = T$_B$* = 750 [°C]

Asimismo, los autores presentan una simulación numérica indicando que el espacio de tiempo entre el perfil inicial (Th) y el actual (Bh-Tl) es de 7000 años, es decir, que la transición del regimen de transferencia de calor de convectivo a conductivo probablemente ocurrió hace 7000 años. A partir de este punto el gradiente geotermico decrece progresivamente, conforme aumenta el tiempo, hasta alcanzar el régimen de estado estacionario bajo pura transferencia de calor conductiva, a los 250, 000 años.

## Modelo Computacional

Para el modelo se empleara una aporximacion por diferencias finitas

$D_{th}\frac{\partial T^2}{\partial z^2} = \frac{\partial T}{\partial t}$

$\frac{d}{dx}(D_t$$_h\frac{dT}{dx}) = \frac{\partial T}{\partial t}$

Donde:

$D_t$$_h = \frac{k(x)}{\rho c}$

```python
# Librerias a emplear
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fdm_1D import Laplaciano1D_NS, calcDth, interpTemp

# Funcion a emplear para

def solNum(L, N, Dth_data, Ttop, Tbot, zfor, Tfor, zcur, Tcur, NT):
    """Calculo de la solucion numerica
    
    Requiere de modulo fdm_1D : Laplaciano 1D, calcDth, interpTemp
    
    Parameters
    ----------------------------------
    L : float
    Profundidad maxima [m]
    
    N : int
    Tamaño de la malla de dominio
    
    Dth_data: ndarray
    Arreglo de datos con parametros en funcion del tiempo
    depth1, depth2, k, cp, rho
    
    Ttop : float
    Temperatura en superficie [°C]
    
    Tbot : float
    Temperatura en superficie [°C]
    
    zfor :
    
    Tfor :
    
    zcur :
    
    Tfor :
    
    NT : int
    Total de Pasos de tiempo [años]

    Returns
    ----------------------------------   
    z : ndarray
    Cordenada de profundidad [m]
    
    Dth : ndarray
    Coeficiente de Difusividad [m2/s]
        
    T : ndarray
    Vector de Temperatura [°C]
    """
    # Datos para la solucion de diferencias finitas
    dx = L / (N+1)              # Distancia entre nodos
    z = np.linspace(0, L, N+2)  # Coordenada z de la profundidad
    dt = 3600*24*365            # Paso de tiempo (seconds per year)
        
    # Aceleración para el método implícito
    # Se emplea un factor de aceleración debido al método implícito
    SP = 1000 # Factor de aceleracion
    dt *= SP
    NT /= SP
    
    Dth = calcDth(Dth_data, z, N) # Vector con el valor del coeficiente Dth(z)
    # Interpolación a los lugares de la simulación
    T_ini = interpTemp(dataTf['zfor'], dataTf['Tfor'], z) # T inicial interpolada
    T_final = interpTemp(dataTc['zcur'], dataTc['Tcur'], z) #T final interpolada
    
    # Definición del sistema lineal de N x N
    dt_dx2 = dt / dx**2
    gamma = Dth * dt_dx2
    A = Laplaciano1D_NS(N, gamma) # Matriz del sistema

    # Ciclo de solución temporal
    #
    T_new = T_ini.copy()
    T_now = np.zeros(N+2)
    T_1 = np.zeros(N+2)
    T_5 = np.zeros(N+2)
    T_9 = np.zeros(N+2)
    T_25 = np.zeros(N+2)
    T_100 = np.zeros(N+2)
    T_250 = np.zeros(N+2)
    for i in range(0, int(NT)+1):
        #print(i)
        T_new[ 1] += gamma[0] * T_ini[0]
        T_new[-2] += gamma[-1] * T_ini[-1]    
        T_new[1:-1] = np.linalg.solve(A,T_new[1:-1])
        if (i == 7):
            T_now = T_new.copy() # Temperatura actual
        elif (i == 1):
            T_1 = T_new.copy() # Temperatura 1,000[años]
        elif (i == 5):
            T_5 = T_new.copy() # Temperatura 5,000[años]
        elif (i == 9):
            T_9 = T_new.copy() # Temperatura 9,000[años]
        elif (i == 25):
            T_25 = T_new.copy() # Temperatura 25,000[años]
        elif (i == 100):
            T_100 = T_new.copy() # Temperatura 100,000[años]
        elif (i == 250):
            T_250 = T_new.copy() # Temperatura 250,000[años]
    
            
    # GRAFICACIÓN
    #
    # Grafico con temperaturas "former" y "current", asi como 
    # la simulacion de la geoterma actual y la simulación final en 250,0000y
    plt.figure ( figsize =(15, 7.5))
    ax1 = plt.subplot(131)
    plt.plot([T_ini[0], T_ini[-1]], [z[0], z[-1]], '--', lw=0.75, c='k' )
    plt.scatter(dataTf['Tfor'], dataTf['zfor'], s=20, alpha=0.5,label='TH data')
    plt.plot(T_ini, z, '-', lw = 1.5, label='Cond. inicial')
    plt.scatter(dataTc['Tcur'], dataTc['zcur'], s=20, alpha=0.5, label='BH-Tl')
    plt.plot(T_now, z, label='Sim. Temp. actual')
    plt.plot(T_new, z, label='Sim. Temp. final')
    plt.ylim(4100,-100)
    plt.xlim(-20,800)
    plt.title('Solución Numérica')
    plt.xlabel('$T$ [$^o$C]')
    plt.ylabel('Depth [m]')
    plt.suptitle('Para N ={:3d}'.format(N))
    plt.grid()
    plt.legend()
    
    # Grafico de la simulación en función del tiempo, con temperaturas "former" (ini)
    # "current" (7,000y) y diferentes pasos de tiempo hasta los 250,000y
    ax2 = plt.subplot(132, sharey=ax1)
    plt.plot(T_ini, z, '--', lw = 1.5, label='Cond. inicial')
    plt.plot(T_now, z,'--', lw = 1.5, label='Sim. Temp. actual')
    plt.plot(T_1, z,'-', lw = 0.5, label='1,000y')
    plt.plot(T_5, z,'-', lw = 0.5, label='5,000y')
    plt.plot(T_9, z,'-', lw = 0.5, label='9,000y')
    plt.plot(T_25, z,'-', lw = 0.5, label='25,000y')
    plt.plot(T_100, z,'-', lw = 0.5, label='100,000y')
    plt.plot(T_250, z,'-', lw = 0.5, label='250,000y')
    plt.ylim(4100,-100)
    plt.xlim(-20,800)
    plt.title('Solución Numérica en función del tiempo')
    plt.xlabel('$T$ [$^o$C]')
    plt.grid()
    plt.legend()
    
    # Grafico de Dth en función de la profundidad
    ax3 = plt.subplot(133, sharey=ax1)
    plt.plot(Dth, z, 'k.--', lw=0.75)
    plt.xlabel('$D_{th}$ [m$^2$/s]')
    plt.xlim(0.65e-6,1.05e-6)
    plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
    plt.title('Coeficiente $D_{th}$')
    plt.grid()

    plt.subplots_adjust(hspace=1.0)

    plt.show()
    
    return z, Dth, T_new

    
# Datos físicos del problema
#
Ttop = 15.   # Temperatura en la superficie
Tbot = 750.  # Temperatura a 4000 m
L = 4000.0  # Profundidad máxima [m]
data = pd.read_csv('phys_dat.csv') # Lectura de datos: k, cp, rho
Dth_data = data['k'] / (data['rho']*data['cp']) # Cálculo del coeficiente Dth
dataTf = pd.read_csv('Form_Temp.csv') # Lectura de datos: zfor,Tfor
dataTc = pd.read_csv('Curr_Temp.csv') # Lectura de datos: zcur,Tcur

print('\nDatos del problema \n{}'.format(data))
print('\n Dth : \n{}'.format(Dth_data))


# Datos para la solución numérica por diferencias finitas
#
N = [10, 25, 50, 100, 250, 500, 1000]
#N = 10
NT = 250000 # Pasos de tiempo

# SOLUCIÓN USANDO DIFERENCIAS FINITAS
#
# Solución del sistema lineal
for n in N:
    # Solucion numerica
    z, Dth, T = solNum(L, n, Dth_data, Ttop, Tbot, dataTf['zfor'], dataTf['Tfor'], dataTc['zcur'], dataTc['Tcur'], NT)
    
#z, Dth, T = solNum(L, N, Dth_data, Ttop, Tbot, dataTf['zfor'], dataTf['Tfor'], dataTc['zcur'], dataTc['Tcur'], NT)
    
```
